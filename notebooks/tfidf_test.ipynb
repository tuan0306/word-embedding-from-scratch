```python
corpus=load_corpus('../data/processed/toy_corpus.txt')
corpus
```

```python
model=TFIDF_Model(corpus)
```

```python
query_1 = ["phòng_trọ", "máy_lạnh"]
print(f"\n>> Khách hàng tìm: {query_1}")
results_1 = model.search(query_1, top=2)
for rank, (idx, score) in enumerate(results_1):
    print(f"Top {rank+1} (Độ khớp {score*100:.1f}%): {' '.join(corpus[idx])}")
```
